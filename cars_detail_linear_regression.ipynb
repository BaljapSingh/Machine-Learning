```python
CompanyName=cars['CarName'].apply(lambda x : x.split(' ')[0])  #to split the carname column
# here we are splicting the values in company name using splict function such that after every starting name of company's name we have space[' '] so from that we splict that total according to starting name of each car
#lambda is a may to create a function it automatically defines the function

cars.insert(3,"CompanyName",CompanyName) # inserting the CompanyName at index 3

cars.drop(['CarName'],axis=1,inplace=True)# here we are droping the carName column (if we donot do this we will have both carName and CompanyName columns)

cars.head(10)
```

```python
print(cars)
```

```python
def replace_name(a,b):
    cars.CompanyName.replace(a,b,inplace=True)
    
replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen ')
replace_name('vw','volkswagen')
cars.tail(30)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(cars[['wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio','stroke','compressionratio','peakrpm','citympg','highwaympg']],cars.price,test_size=0.20)
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_predict=reg.predict(x_test)
y_predict
```
